# Création de la colonne mot vedette

```python
#import de librairie
import pandas as pd
```

```python
# chemins des deux fichiers à lire 
cheminDuFichierCSV = "./../data/dictionary/Buchanan_PronChar_counts.csv" 
CheminCSV2="./../data/dictionary/1757_Buchanan-J.csv" 
# créations des dataframes pandas correspondant 
dfmapping = (pd.read_csv(cheminDuFichierCSV, delimiter=";", encoding="utf8"))
dfDicoBuch = pd.read_csv(CheminCSV2, delimiter=";", encoding="utf8")
#remplissage des valeur nulles / non valide par un vide de chaine de caractère 
dfmapping['graph']=dfmapping['graph'].fillna("")
```

Affichage des deux fichiers CSV que l'on vas traité.

```python
dfDicoBuch
```

```python
dfmapping
```

```python
#création de la fonction Buch2Class
def buch2Class(strBuch, dfMapping):
    #création de la variable résultat
    BuchRes = ""
    listValBuch = strBuch
    # pour un mot particuler en Buch
    for valBuch in listValBuch:
        #on regarde si on trouve une correspondance dans notre table de mapping
        dfRowFinded = dfMapping[dfMapping['char']==valBuch]
        #si on en trouve bien une alors 
        if len(dfRowFinded==1):
            #on ajoute a notre mot le caractère corespondant
            BuchRes+=dfRowFinded.iloc[0]['graph']
        else:
            BuchRes+='?'
    return BuchRes
```

On effectue une mannipulation afin d'avoir le mot Buch traduit en un mot classique

```python
%%time
dfDicoBuch['motVedette'] = dfDicoBuch.apply(lambda x: buch2Class(x['pron'], dfmapping),1)
```

```python
dfDicoBuch
```

```python
#on enregistre notre dictionaire (sauvegarde de secours)
dfDicoBuch.to_csv("./../data/dataframe/dfDicoBuch.csv",index=False)
```

# Création des colonnes Sampa et Ipa 

```python
# récupére notre dico enregistrer précédement.  
dfDicoBuch = pd.read_csv("./../data/dataframe/dfDicoBuch.csv" , delimiter=",", encoding="utf8")
```

```python
#on récupère un autre dictionnaire  nommé Bigi
cheminDuDicoBigi = "./../data/dataframe/BigiRichSample.csv" 
dfBigi = (pd.read_csv(cheminDuDicoBigi, delimiter=";", encoding="utf8"))
```

Afin de faire la création des colonnes Sampa et Ipa on utilise le fichier BigiRichSample.csv

```python
dfDicoBuch
```

```python
dfBigi
```

```python
#même principe que pour la fonction Buch2Class. 
#cet fois si on cherche basiquement si le mot ce trouve dans la colonne Sampa ou Ipa du dico de Bigi. sans chercher de caractère particulier. 
def addSampa(strBuch, dfBigi):
    BuchRes = ""
    dfRowFinded = dfBigi[dfBigi['hwd']==strBuch]
    if len(dfRowFinded==1):
        BuchRes=dfRowFinded.iloc[0]['pronSampa']
    else:
        BuchRes='?'
    return BuchRes

def addIpa(strBuch, dfBigi):
    BuchRes = ""
    dfRowFinded = dfBigi[dfBigi['hwd']==strBuch]
    if len(dfRowFinded==1):
        BuchRes=dfRowFinded.iloc[0]['ipa']
    else:
        BuchRes='?'
    return BuchRes
```

```python
dfDicoBuch['Sampa'] = dfDicoBuch.apply(lambda x: addSampa(x['motVedette'], dfBigi),1)
dfDicoBuch['Ipa'] = dfDicoBuch.apply(lambda x: addIpa(x['motVedette'], dfBigi),1)
```
